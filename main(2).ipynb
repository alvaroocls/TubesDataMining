Variable target dataset ini tidak seimbang (imbalanced), dengan jumlah data kategori <=50K lebih banyak dibandingkan >50K.
Ketidakseimbangan ini perlu diperhatikan dalam proses analisis atau pelatihan model machine learning, karena dapat memengaruhi performa model, khususnya pada prediksi kategori minoritas (>50K).

```python
def countplot_on_income_grid_as_ratio(df):
    # Mendapatkan semua kolom bertipe 'object' (kecuali 'income' jika ada)
    object_columns = [col for col in df.select_dtypes(include='object').columns if col != 'income']
    
    n_cols = 3  # Menentukan jumlah kolom dalam grid
    n_rows = (len(object_columns) + n_cols - 1) // n_cols  # Menghitung jumlah baris yang dibutuhkan
    
    # Mengatur ukuran figure
    plt.figure(figsize=(n_cols * 5, n_rows * 5))
    
    # Loop melalui semua kolom bertipe 'object' dan buat barplot untuk rasio masing-masing
    for index, col in enumerate(object_columns):
        # Menghitung proporsi setiap kategori dalam kolom berdasarkan 'income'
        prop_df = df.groupby([col, 'income']).size().reset_index(name='count')
        total_df = df.groupby(col).size().reset_index(name='total')
        merged_df = pd.merge(prop_df, total_df, on=col)
        merged_df['ratio'] = merged_df['count'] / merged_df['total']
        
        plt.subplot(n_rows, n_cols, index + 1)  # Menentukan posisi subplot
        sns.barplot(x=col, y='ratio', hue='income', data=merged_df)
        plt.xticks(rotation=90)
        plt.title(col)
        plt.ylabel('Ratio')
        plt.tight_layout()

    plt.show()

# Contoh penggunaan
countplot_on_income_grid_as_ratio(df)
```

## 2. Data Preparation

### ‚Ä¢ Map target variable menjadi 0 dan 1

```python
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})
```

### ‚Ä¢ Handle baris yang null

Ganti row yang memiliki '?' menjadi nan agar bisa dilakukan pemrosesan lebih lanjut. Preprocessing steps ini dilakukan untuk menangani missing values dalam dataset dengan mengisi nilai yang hilang menggunakan modus dari kolom yang bersangkutan. Tujuan dari langkah-langkah ini adalah memastikan tidak ada missing values dalam dataset yang bisa mengganggu proses pelatihan model machine learning. 

```python
df.replace('?', np.nan, inplace=True)
df.isnull().sum()
```

```python
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)
df['occupation'].fillna(df['occupation'].mode()[0], inplace=True)
df['native.country'].fillna(df['native.country'].mode()[0], inplace=True)
```

## 3. Data Modeling & Evaluation

### ‚Ä¢ Train Test Split

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='income')
y = df['income'].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### ‚Ä¢ Fungsi Evaluasi Model

```python
from sklearn.metrics import f1_score,confusion_matrix,classification_report,accuracy_score


def train_and_evaluate_model(model, model_name,X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    
    predictions_test = model.predict(X_test)
   
    test_accuracy_f1 = f1_score(y_test, predictions_test) * 100
    
    
    print(f"{model_name} F1 Score Testing -> {test_accuracy_f1:.2f}%")
    print(f"{model_name} Accuracy Score: -> {(accuracy_score(y_test, predictions_test) *100):.2f}%")
     
    print(f"{model_name} Confusion Matrix:")
    cm = confusion_matrix(y_test, predictions_test)
    print(cm)

    print(f"{model_name} Classification Report:")
    print(classification_report(y_test, predictions_test))

    from sklearn.metrics import roc_auc_score
    print(f"{model_name} ROC AUC Score: -> {(roc_auc_score(y_test, predictions_test) *100):.2f}%")
    
    print('-' * 50) 
    return model
```

### ‚Ä¢ Model Klasifikasi

```python

from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

cb_modelweight = train_and_evaluate_model(CatBoostClassifier(verbose=0,class_weights=[1,1.5]), 'CatBoost',X_train, X_test, y_train, y_test)
xgb_modelweight = train_and_evaluate_model(XGBClassifier(scale_pos_weight = 1.5), 'XGBoost',X_train, X_test, y_train, y_test)
lr_modelweight = train_and_evaluate_model(LogisticRegression(class_weight={0:1,1:1.5}), 'Logistic Regression',X_train, X_test, y_train, y_test)
```

### Hyperparameter tuning

parameter tuning for xgboost

```python

from sklearn.model_selection import GridSearchCV

param_grid = {
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'scale_pos_weight': [1.25, 1.5, 1.75]
}

xgb = XGBClassifier()
grid_search_xgb = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2,scoring='f1')
grid_search_xgb.fit(X_train, y_train)

grid_search_xgb.best_params_,grid_search_xgb.best_score_
```

Parameter tuning for catboost

```python
param_grid = {
    'depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'iterations': [100, 200, 300],
    'scale_pos_weight': [1.25, 1.5, 1.75]
}

cb = CatBoostClassifier(verbose=0)

grid_search_cb = GridSearchCV(estimator=cb, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2,scoring='f1')
grid_search_cb.fit(X_train, y_train)

grid_search_cb.best_params_,grid_search_cb.best_score_
```

Parameter tuning for logistic regression

```python
param_grid = {
    'C': [0.1, 1, 10],
    'max_iter': [100, 200, 300],
    'class_weight': [{0:1,1:1.25},{0:1,1:1.5},{0:1,1:1.75}]
}

lr = LogisticRegression()

grid_search_lr = GridSearchCV(estimator=lr, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2,scoring='f1')
grid_search_lr.fit(X_train, y_train)

grid_search_lr.best_params_,grid_search_lr.best_score_
```

Evaluate model after tuning

```python
cb_model_tuned = train_and_evaluate_model(CatBoostClassifier(**grid_search_cb.best_params_), 'CatBoost',X_train, X_test, y_train, y_test)
xgb_model_tuned = train_and_evaluate_model(XGBClassifier(**grid_search_xgb.best_params_), 'XGBoost',X_train, X_test, y_train, y_test)
lr_model_tuned = train_and_evaluate_model(LogisticRegression(**grid_search_lr.best_params_), 'Logistic Regression',X_train, X_test, y_train, y_test)
```

### Clustering Model

```python
df_clustering = df.copy()
df_clustering
```

Feature selection

```python
#pick features for clustering
df_clustering = df_clustering[['workclass','capital','education.num']]
```

```python
#pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_clustering)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
#elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(principalDf)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')

plt.show()
```

```python
#kmeans with 2 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(principalDf)
principalDf['cluster'] = kmeans.labels_
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x='principal component 1', y='principal component 2', hue='cluster', data=principalDf, palette='viridis')
plt.title('Clusters based on PCA')
plt.show()
```

```python
#evaluate clustering
from sklearn.metrics import silhouette_score
silhouette_score(principalDf, kmeans.labels_)
```
